# Pet Recommendation System

In this kernel we'll be building a Pet Recommendation System using https://github.com/sahityashetty19/pet-recommender-system/tree/master/data Dataset using <b> K-Means clustering using Manhattan Distance for ordinal/likert scale data </b>. https://medium.com/analytics-vidhya/customer-segmentation-using-k-prototypes-algorithm-in-python-aad4acbaaede

The user is able to input the name of a dog breed (as a string), for which the system will have to return a set of the breeds that are most similar to this breed based on a specific attribute (excluding itself). To aid this the dataset will contain a set of dictionary dogs, mapping to dog breeds to various adgectives describing that breed's typical attributes such as its height, weight, size. In this context, two breeds are more similar the more attributes they share.

# Model Training

```python
dataset.head(5)
```

```python
pet_array = dataset.values
pet_array = pet_array[:, :-1]
```

```python
# Define a custom Manhattan distance function
def manhattan_distance(x, y):
    return sum(abs(a - b) for a, b in zip(x, y))
```

```python
def plot_elbow_curve(start, end, data):
    no_of_clusters = list(range(start, end+1))
    cost_values = []

    for k in no_of_clusters:
        # Create a custom Manhattan distance metric
        manhattan_metric = distance_metric(type_metric.MANHATTAN)

        # Initialize cluster centers using k-means++
        initial_centers = kmeans_plusplus_initializer(data, k).initialize()

        # Create K-Means algorithm with custom distance metric
        kmeans_instance = kmeans(data, initial_centers, metric=manhattan_metric)

        # Run the K-Means algorithm
        kmeans_instance.process()
        
        # Get the cost (inertia) value
        cost_values.append(kmeans_instance.get_total_wce())

    plt.figure(figsize=(15, 7))
    ax = sns.lineplot(x=no_of_clusters, y=cost_values, marker="o", dashes=False)
    ax.set_title('Elbow curve', fontsize=18)
    ax.set_xlabel('No of clusters', fontsize=14)
    ax.set_ylabel('Cost', fontsize=14)
    ax.set(xlim=(start-0.1, end+0.1))
    plt.plot();
    
# Plotting elbow curve for k=2 to k=10
plot_elbow_curve(2,10,pet_array)
```

```python
# Number of clusters (you can adjust this)
k = 6

# Create a custom Manhattan distance metric
manhattan_metric = distance_metric(type_metric.MANHATTAN)

# Initialize cluster centers using k-means++
initial_centers = kmeans_plusplus_initializer(pet_array, k).initialize()

# Create K-Means algorithm with custom distance metric
kmeans_instance = kmeans(pet_array, initial_centers, metric=manhattan_metric)

# Run the K-Means algorithm
kmeans_instance.process()

# Get cluster assignments and cluster centers
clusters = kmeans_instance.get_clusters()
cluster_centers = kmeans_instance.get_centers()

# Print cluster assignments
for cluster_idx, cluster in enumerate(clusters):
    print(f"Cluster {cluster_idx}: {cluster}")
```

```python
dataset['cluster'] = None
for cluster_idx, cluster in enumerate(clusters):
    for j in cluster:
        dataset.loc[dataset.index[j], 'cluster'] = f'{cluster_idx}'
```
